```python
#importa as bibliotecas
import numpy as np, scipy as sp, matplotlib.pyplot as plt
from scipy import signal

# Definindo FUNÇÃO DE TRANSFERÊNCIA
tau = 10
num = [2, 1, 2]
den = [tau, 1, +.9]
sys = signal.lti(num, den)

# Resposta ao impulso
tempo, yout = signal.impulse(sys)

# Plot da resp ao impulso
plt.figure(figsize=(8,3))
plt.plot(tempo, yout)
plt.title("Resposta ao impulso")
plt.xlabel("tempo")
plt.ylabel("$h(t)$");
```

```python
# Resposta ao degrau
t_degrau, y_degrau = signal.step(sys)

# Plot da resp ao impulso
plt.figure(figsize=(8,3))
plt.plot(t_degrau, y_degrau)
plt.title("Resposta ao degrau")
plt.xlabel("tempo")
plt.ylabel("$y(t)$");
```

```python
# Diagrama de Bode
w = np.logspace(-3, 2, 100)
w, mag, phase = signal.bode(sys, w=w)

# Módulo
plt.figure(figsize=(15,3))
plt.semilogx(w, mag)
plt.title("Diagrama de Bode")
plt.xlabel("$\omega$")
plt.ylabel("$|H(j\omega)|$ (dB)")
plt.grid(1);

# Fase
plt.figure(figsize=(15,3))
plt.semilogx(w, phase)
plt.xlabel("$\omega$")
plt.ylabel("$\measuredangle H(j\omega)$");
plt.grid(1)
plt.yticks(np.arange(0,-120,-10));
```

```python
# Resposta em frequência
w, H = signal.freqresp(sys)

# Módulo
plt.figure(figsize=(15,3))
plt.plot(w, np.abs(H))
plt.title("Resposta em Frequencia")
plt.xlabel("$\omega$")
plt.ylabel("$|H(j\omega)|$")
plt.grid(1);

# Fase
plt.figure(figsize=(15,3))
plt.plot(w, np.angle(H))
plt.xlabel("$\omega$")
plt.ylabel("$\measuredangle H(j\omega)$");
plt.grid(1);
```
