```python
#importa as bibliotecas
import numpy as np, scipy as sp, matplotlib.pyplot as plt
from scipy import signal

# parametros
K = 1
tau = 2
num = [K]
den = [tau, 1]

# sistema dinamico
sys = signal.lti(num, den)
# resposta ao impulso
t, y = signal.impulse(sys)

plt.figure()
plt.plot(t, y)
plt.xlabel('tempo (s)')
plt.ylabel('$h(t)$');

# parametros
K = 1
tau = 4
num = [K]
den = [tau, 1]

# sistema dinamico
sys2 = signal.lti(num, den)
# resposta ao impulso
t, y2 = signal.impulse(sys2)

plt.plot(t, y2);
plt.title("Resposta ao impulso");
```

```python
w1, H1 = signal.freqresp(sys)
w2, H2 = signal.freqresp(sys2, w1)

plt.figure()
plt.plot(w1, np.abs(H1)) #modulo de H1
plt.plot(w2, np.abs(H2)) #modulo de H2
plt.xlabel("frequencias (rad/s)")
plt.ylabel("Modulo")
plt.title("Resposta em Frequencia");
```

```python
t1, y1 = signal.step(sys)
t2, y2 = signal.step(sys2)

plt.figure()
plt.plot(t1, y1)
plt.plot(t2, y2)
plt.xlabel("tempo (s)")
plt.ylabel("Resposta ao degrau")
plt.title("Resposta ao degrau");
```

```python
w1, mag1, phase1 = signal.bode(sys)
w2, mag2, phase2 = signal.bode(sys2, w1)

plt.figure()
plt.semilogx(w1, mag1) #modulo de H1
plt.semilogx(w2, mag2) #modulo de H2
plt.xlabel("frequencias (rad/s)")
plt.ylabel("Modulo")
plt.title("Resposta em Frequencia")
plt.grid(1);
```

```python
t = np.arange(0, 100, .1)
u = np.heaviside(t-10, 0) - np.heaviside(t-30, 0)

t1, y1, _ = signal.lsim(sys, u, t)
t2, y2, _ = signal.lsim(sys2, u, t)

plt.figure()
plt.plot(t, u)
plt.plot(t1, y1)
plt.plot(t2, y2)
plt.xlabel("tempo (s)")
plt.ylabel("$y(t)$")
plt.title("Resposta a um sinal de entrada");
```
