```python
# Importa bibliotecas
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import IPython.display as ipd
from scipy.io import wavfile
from scipy import signal
# para figuras estáticas, use:
# %matplotlib inline
# OU, para figuras interativas, use:
%matplotlib widget
```

```python
# lê arquivo de áudio
fs, data = wavfile.read('hallelujah.wav')
data = (data-np.mean(data))/255
tempo = np.arange(len(data))/fs*1000

plt.figure(figsize=(8,1.5))
# plt.plot(tempo, data)
# plt.plot(tempo[::3], data[::3]) #sinal manipulado
plt.plot(data)
plt.plot(data[::3]-1) #sinal manipulado
plt.xlabel("tempo [ms]")
# plt.xlim((0,max(tempo)))
plt.show();
```

```python
# player para o arquivo original
aleluia = data
ipd.Audio(aleluia, rate=fs, normalize=False)
```

### Alteração na escala vertical

```python
# audio alterado na escala vertical
aleluia = 0.3*data
ipd.Audio(aleluia, rate=fs, normalize=False)
```

### Inversão no tempo

```python
# audio com inversão temporal (e alteração na escala vertical)
aleluia = data[::-1]
ipd.Audio(aleluia, rate=fs, normalize=False)
```

### Compressão no tempo

```python
# compressão no tempo
aleluia = data[::3]
ipd.Audio(aleluia, rate=fs, normalize=False)
```

### Deslocamento temporal

```python
# deslocamento temporal (truncamento das amostras iniciais)
aleluia = data[70000::]
ipd.Audio(aleluia, rate=fs, normalize=False)
```

### Espectrograma

```python
f, t, Sxx = sp.signal.spectrogram(data, fs)
plt.figure(figsize=(12,3))
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
